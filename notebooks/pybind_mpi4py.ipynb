```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

%matplotlib inline
```

## Seidel method

```python
pattern = '../build/vis/result_matrix_*.txt'
files = glob.glob(pattern)

fig, ax = plt.subplots()

def update(frame):
    
    ax.clear()
    
    file = f'../build/vis/result_matrix_{frame}.txt'
    array = np.loadtxt(file)
    
    plot = ax.imshow(array)
    return plot, 

ani = FuncAnimation(fig, update, frames=np.arange(0, len(files) // 2), repeat=True)
ani.save('ani.gif', writer='pillow', fps=15)
```
